# `Scatter plot`

```python
import matplotlib.pyplot as plt
import numpy as np

share = [85, 50, 75, 80, 34, 67, 65]
price = [55, 44, 48, 52, 43, 89, 66]
colors = [10, 49, 30, 29, 56]

plt.title('Scatter plot', fontsize=20)
plt.xlabel('share', fontsize=15, c='b')
plt.ylabel('price', fontsize=15, c='b')
plt.scatter(share, price, c='r', marker='*', cmap='viridis')
plt.colorbar()
plt.show()
```

# `Pichart`

```python
import matplotlib.pyplot as plt
x = [10, 15, 20, 25, 30]
y = ['c', 'c++', 'java', 'javascript', 'python']
c = ['r', 'b', 'y', 'g', 'o']
plt.pie(x, labels=y, autopct='%0.1f%%', radius=1, labeldistance=1.2, startangle=45, counterclock=False)
plt.legend(loc=3)
plt.show()
```

# `Stem plot`

# `Step Plot`

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 5, 4]

plt.step(x, y)
plt.title('Pytorch')
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.grid()
plt.show()
```

# `Fill Between`

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])

# Change 'colors' to 'color'
plt.plot(x, y, color='r')
plt.fill_between(x, y, color='g', where=(x>=2) & (x<=4), alpha=1) # Change 'colors' to 'color'
plt.title('Pytorch')
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.show()
```

# `Subplot Plot`

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 2, 3, 4]

# display plot
# plt.plot(x, y, color='r')

m = [1, 2, 3, 4]
n = [3, 1, 5, 2]
plt.plot(m, n, color='b')
plt.xticks(m, labels=['python', 'c', 'c++', 'java'])
plt.yticks(x)

# using limit function
plt.xlim(0, 10)
plt.ylim(0, 10)

plt.title('coding race', fontsize=15)
plt.xlabel('popularity', fontsize=15)
plt.ylabel('trends', fontsize=15)
plt.text(2, 5, 'java', fontsize=15)


plt.show()

```
